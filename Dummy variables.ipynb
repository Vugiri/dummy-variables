```python
import pandas as pd
```

```python
df=pd.read_csv("carprices.csv")
df
```

```python
dummies=pd.get_dummies(df.Car)
dummies
```

```python
conc=pd.concat([df,dummies],axis='columns')
```

```python
conc
```

```python
k=conc.drop(['Mercedez Benz C class','Car'],axis='columns')
```

```python
k
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
X=k.drop(['Mileage'],axis='columns')
```

```python
X
```

```python
y=k.Mileage
```

```python
y
```

```python
model.fit(X,y)
```

```python
model.predict([[18000,6,0,1]])
```

```python
model.score(X,y)
```
